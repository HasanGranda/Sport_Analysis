```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns
```

```python
#Messi vs Real Betis veri okuma
df = pd.read_csv('messibetis.csv')
```

```python
df['x'] = df['x']*1.2
df['y'] = df['y']*.8
df['endX'] = df['endX']*1.2
df['endY'] = df['endY']*.8
```

```python
df
```

```python
fig ,ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

#Saha oluşturma 'statsbomb'
pitch = Pitch(pitch_type='statsbomb', orientation='horizontal',
              pitch_color='#22312b', line_color='#c7d5cc', figsize=(13, 8),
              constrained_layout=False, tight_layout=True)

#Saha çizimi
pitch.draw(ax=ax)
plt.gca().invert_yaxis()

#Isı haritası oluşturma
kde = sns.kdeplot(
        df['x'],
        df['y'],
        shade = True,
        thresh=0.1,
        alpha=.5,
        n_levels=10,
        cmap = 'magma'
)


#Her pas için loop
for x in range(len(df['x'])):
    if df['outcome'][x] == 'Successful':
        plt.plot((df['x'][x],df['endX'][x]),(df['y'][x],df['endY'][x]),color='green')
        plt.scatter(df['x'][x],df['y'][x],color='green')
    if df['outcome'][x] == 'Unsuccessful':
        plt.plot((df['x'][x],df['endX'][x]),(df['y'][x],df['endY'][x]),color='red')
        plt.scatter(df['x'][x],df['y'][x],color='red')
        
plt.xlim(0,120)
plt.ylim(0,80)

plt.title('Messi Pas Haritası vs Real Betis',color='white',size=20)
plt.savefig('Messi Pas Haritası.png')
```
